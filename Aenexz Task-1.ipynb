```python
# find the largest of three numbers.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Largest number is", a)
elif b > c:
    print("Largest number is", b)
else:
    print("Largest number is", c)
```

```python
# check whether a given year is a leap year
year = int(input("Enter a year: "))

if year % 4 == 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
# check if a character is a vowel or consonant
char = input("Enter a character: ")
if char in 'aeiou':
     print("Vowel")
else:
    print("Consonant")
```

```python
# check whether a number is divisible by both 5 and 11
num = int(input("Enter a number: "))

if num % 5 == 0 and num % 11 == 0:
    print("The number is divisible by both 5 and 11")
else:
    print("The number is not divisible by both 5 and 11")
```

```python
# calculate the sum of first N natural numbers using a while loop
n = int(input("Enter a number: "))
sum = 0
i = 1
while i <= n:
    sum = sum + i
    i = i + 1
print("Sum =", sum)
```

```python
# print the multiplication table of a given number using a while loop
num = int(input("Enter a number: "))
i = 1
while i <= 10:
    print(num, "x", i, "=", num * i)
    i = i + 1
```

```python
# Pyramid Pattern
rows = 4
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```

```python
# Number Pattern Program
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```
