```python
# function that takes the radius of a circle and returns its area
def area_of_circle(radius):
    area = 3.14 * radius * radius
    return area
r = float(input("Enter radius: "))
print("Area of circle =", area_of_circle(r))
```

```python
# function that accepts three numbers and returns the largest one
def largest(a, b, c):
    if a > b and a > c:
        return a
    elif b > c:
        return b
    else:
        return c
print(largest(10, 25, 15))
```

```python
# function that takes a string and returns its length without using built-in length functions
def string_length(text):
    count = 0
    for ch in text:
        count = count + 1
    return count
print(string_length("Python"))
```

```python
# function that checks whether a number is prime
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True
n = int(input("Enter a number: "))

if is_prime(n):
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
# recursive function to find the nth Fibonacci number
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
n = int(input("Enter n: "))
print(fibonacci(n))
```

```python
# recursive function to reverse a string
def reverse_string(text):
    if text == "":
        return ""
    else:
        return reverse_string(text[1:]) + text[0]
print(reverse_string("Giridhar"))
```

```python
# program to find the sum of all numbers from 1 to n using a for loop
n = int(input("Enter a number: "))
sum = 0
for i in range(1, n + 1):
    sum = sum + i
print("Sum =", sum)
```

```python
# program to count how many even numbers are present in a list
numbers = [10, 15, 20, 25, 30, 35]
count = 0
for num in numbers:
    if num % 2 == 0:
        count = count + 1
print("Number of even numbers =", count)
```

```python
# program to print all prime numbers between 1 and 100
for num in range(2, 99):
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    if prime:
        print(num)
```
